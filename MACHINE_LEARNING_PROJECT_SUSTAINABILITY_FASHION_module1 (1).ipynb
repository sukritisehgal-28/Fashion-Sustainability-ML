# **FASHION SUSTAINABILITY ANALYSIS**

---

The dataset is sourced from kaggle, represents information in sustainable fashion with several columns of data. The dataset includes carbon footprint, waste generated, water used, price range, and product lines for brands. Secondly, the dataset discusses the materials applied to each type of apparel, the sustainability rating, eco-friendly manufacturing, and recycling facilities. The dataset shall help establish a trend in regard to sustainable fashion through analysis that shows how different brands apply their concept to make their products sustainable. The project will entail in-depth analysis of these factors toward informing insight into the environmental impact of the fashion industry, showing areas that can be worked upon.


KNOWLEDGE ABOUT THE DATASET



*   Brand_Name: Name of the fashion brand
*   Country: Country where the brand is based
* Sustainable Material (%): Percentage of sustainable materials used by
 the brand
* Production Process: Method used for producing the clothing (e.g., recycled, organic)
* Consumer Engagement Score: A score (out of 10) reflecting the brand’s impact on sustainable consumer behavior
* Year: Year the data was collected
* Sustainability_Rating: The sustainability ratings of the brands.
* Material_Type: Different types of materials used by the brands.
* Eco_Friendly_Manufacturing: Whether brands adopt eco-friendly manufacturing processes.

* Carbon_Footprint_MT: Distribution of carbon footprints in metric tons.
* Water_Usage_Liters: Water usage of the brands in liters.

* Waste_Production_KG: Distribution of waste production by brands in kilograms.
* Recycling_Programs: Availability of recycling programs among the brands.
* Product_Lines: Variety of product lines offered by the brands.
* Average_Price_USD: Distribution of average prices in USD for each brand.
* Market_Trend: Market trends across the dataset.
* Certifications: Different sustainable model awardee certifications held by brands.

```python
#counting the outliners for all the numeric coloumns
def count_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers)

total_outliers = 0
for col in ['Year', 'Average_Price_USD', 'Carbon_Footprint_MT', 'Water_Usage_Liters', 'Waste_Production_KG']:
    if col in fashion_sustainability_Ds.columns:
        num_outliers = count_outliers_iqr(fashion_sustainability_Ds, col)
        total_outliers += num_outliers
        print(f"Number of outliers in '{col}': {num_outliers}")
    else:
        print(f"Column '{col}' not found in the dataset.")

print(f"\nTotal number of outliers in all the numeric columns: {total_outliers}")
```

```python
fashion_data = pd.read_csv('fashion_sustainability.csv')
fashion_data.head()
```

```python
#numeric columns for outlier detection
num_cols = fashion_data.select_dtypes(include=['float64', 'int64']).columns

Q1 = fashion_data[num_cols].quantile(0.25)
Q3 = fashion_data[num_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
fashion_data_cleaned = fashion_data[~((fashion_data[num_cols] < lower_bound) | (fashion_data[num_cols] > upper_bound)).any(axis=1)]

# Save the cleaned dataset
fashion_data_cleaned.to_csv("fashion_sustainability_cleaned.csv", index=False)
print(f"Outliers removed! Cleaned dataset saved with {fashion_data_cleaned.shape[0]} rows remaining.")
```

```python
#counting the outliners for all the numeric coloumns after removing it
def count_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers)

total_outliers = 0
for col in ['Year', 'Average_Price_USD', 'Carbon_Footprint_MT', 'Water_Usage_Liters', 'Waste_Production_KG']:
    if col in fashion_data_cleaned.columns:
        num_outliers = count_outliers_iqr(fashion_data_cleaned, col)
        total_outliers += num_outliers
        print(f"Number of outliers in '{col}': {num_outliers}")
    else:
        print(f"Column '{col}' not found in the dataset.")

print(f"\nTotal number of outliers in all the numeric columns: {total_outliers}")
```

```python
#carbon footprint over the years
carbon_footprint_by_year = final_data.groupby('Year')['Carbon_Footprint_MT'].mean()


plt.figure(figsize=(10, 6))
plt.plot(carbon_footprint_by_year.index, carbon_footprint_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Carbon Footprint (MT)')
plt.title('Average Carbon Footprint Over the Years')
plt.grid(True)
plt.show()
```

This line plot follows the trend of the average carbon footprint over the years, showing great fluctuation with a general downtrend. Peaks around 2017 show the higher impact that might be related to higher production or unsustainable practice. The overall downtrend reflects improvements in sustainability efforts, such as the adoption of greener technologies or more stringent environmental regulations. The notable dips, such as that in 2022, may correspond to external events like policy changes or reduced activities of production. Further analysis could be done to explore drivers of change for better understanding of the progress toward sustainability.

```python
#correlation heatmap between carbon foot print, water usage, waste production
plt.figure(figsize=(10,6))
material_counts = final_data.groupby(['Material_Type'])[['Carbon_Footprint_MT', 'Water_Usage_Liters', 'Waste_Production_KG']].mean()
sns.heatmap(material_counts.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Material Sustainability Metrics")
plt.show()
```

The heatmap reveals the correlations between carbon footprint, water usage, and waste production in textile materials. There is a moderate positive correlation (0.51) between carbon footprint and water usage, suggesting that materials with a higher carbon footprint tend to consume more water. Conversely, carbon footprint and waste production show a moderate negative correlation (-0.47), indicating that materials with higher carbon footprints may generate less waste. Meanwhile, the correlation between water usage and waste production is negligible (-0.0072), showing no significant relationship. These findings highlight trade-offs among sustainability metrics in textile materials.The heatmap highlights trade-offs in textile sustainability metrics. Materials with a higher carbon footprint tend to consume more water but may produce less waste, while water usage and waste production are largely independent. This suggests that improving one metric may impact others, requiring a balanced approach to sustainability.

```python
#plot between waste production and max water usage per material
sns.set_style("whitegrid")

fig, ax1 = plt.subplots(figsize=(12, 6))

sns.barplot(data=final_data, x='Material_Type', y='Waste_Production_KG', palette='viridis', ax=ax1)
ax1.set_ylabel('Waste Production (KG)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
sns.lineplot(data=final_data, x='Material_Type', y='Water_Usage_Liters', ax=ax2, marker="o", errorbar=None)
ax2.set_ylabel('Water Usage (Liters)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.xticks(rotation=45)
plt.title('Waste Production and Maximum Water Usage by Material Type')
plt.xlabel('Material Type')
plt.show()
```

This graph shows clearly the relation of waste generated with maximum water use for various kinds of textile material. Hemp and Tencel present similar levels of waste, but Tencel is way ahead in water consumption. Bamboo Fabric generates the least waste but has a remarkable peak of water usage. Recycled Polyester has its generation of waste moderated and with relatively low water consumption, proving to be an ecological material. Organic Cotton represents a middle rate for both headings, while Vegan Leather produces more waste but at the same time uses comparatively little water. The insights indicate wastage and resource efficiency trade-off for different types of materials for sustainable material choices.

```python
#boxplot showing wasteproduction based on material type and price category



final_data['Price_Category'] = pd.cut(final_data['Average_Price_USD'], bins=[0, 100, 200, 300, 400, 500], labels=['Low', 'Medium', 'High', 'Very High', 'Premium'])


plt.figure(figsize=(12, 8))
sns.boxplot(x='Material_Type', y='Waste_Production_KG', hue='Price_Category', data=final_data)
plt.title('Boxplot of Waste Production by Material Type and Price Category')
plt.xlabel('Material Type')
plt.ylabel('Waste Production (KG)')
plt.xticks(rotation=45)
plt.legend(title='Price Category')
plt.tight_layout()
plt.show()
```

The boxplot shows waste production across different material types and price categories (Low, Medium, High, Very High, Premium). Waste production remains relatively consistent across materials, with no extreme outliers or distinct patterns. Premium and Very High price categories generally have slightly higher median waste production compared to lower categories. The interquartile ranges (IQRs) are wide, indicating substantial variability within the data. Materials like Bamboo Fabric and Recycled Polyester exhibit tighter groupings of medians, suggesting less variability in waste production across price categories. This highlights that while price and material influence waste production, other factors likely contribute to the observed variability.
